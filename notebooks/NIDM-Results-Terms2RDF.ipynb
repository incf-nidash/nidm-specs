# Extract NIDM-Results Terms and Convert to RDF

- CSV file from Google Spreadsheets with NIDM-Results terms
- Convert terms and definitions to RDF using the RDF branch of PROV Python library
- Organize terms by type - either PROV Entity, Activity, Agent, or Attribute
- For Attributes, map to correct property type - either Object Property, Data Property or Annotation

```python
import os
from uuid import uuid1

import pandas as pd
import prov.model as prov
from rdflib.namespace import RDF, RDFS, XSD
```

```python
df = pd.read_csv("data/NIDM-Results-Terms.csv")
# Replace "NA" by empty cell as suggested in 
# http://stackoverflow.com/questions/10867028/get-pandas-read-csv-to-read-empty-values-as-empty-string-instead-of-nan
df = df.fillna('') 
print(df.info())
```

```python
attrs_filter = df['Prov Type'] == 'Attribute'
attrs = df[attrs_filter]
entity_filter = df['Prov Type'] == 'Entity'
entity = df[entity_filter]
act_filter = df['Prov Type'] == 'Activity'
act = df[act_filter]
```

```python
# Namespace declarations
NIDM = prov.Namespace("nidm", "http://www.incf.org/ns/nidash/nidm#")
SPM = prov.Namespace("spm", "http://www.incf.org/ns/nidash/spm#")
FSL = prov.Namespace("fsl", "http://www.incf.org/ns/nidash/fsl#")
NIIRI = prov.Namespace("niiri", "http://nidm.nidash.org/iri/")
OWL = prov.Namespace('owl', 'http://www.w3.org/2002/07/owl#')
CRYPTO = prov.Namespace('crypto', 'http://id.loc.gov/vocabulary/preservation/cryptographicHashFunctions#')
ProvXSD = prov.Namespace('xsd', 'http://www.w3.org/2001/XMLSchema#')

# Information Artifact Ontology used for "example of usage" term IAO_0000112 and "editor note" IAO_0000116 (as in STATO)
IAO = prov.Namespace('iao', 'http://purl.obolibrary.org/obo/iao.owl#')
```

```python
def updated_namespace(df, termList):
    was_list = True
    
    if not isinstance(termList, list):
        termList = [termList]
        was_list = False;
        
    comments = ""
        
    # For each (domain or attribute) term retreive the correponding namespace (and ignore terms from external namespaces)
    for i, element in enumerate(termList):
        if element.lower() == "integer":
            termList[i] = ProvXSD["int"]
        elif element.lower() == "float":
            termList[i] = ProvXSD["float"]
        elif element.lower() == "true/false":
            termList[i] = ProvXSD["boolean"]
        elif element.lower() == "positiveinteger":
            termList[i] = ProvXSD["positiveInteger"]
        elif element.lower() == "string":
            termList[i] = ProvXSD["string"]
        elif element == "prov:location":
            termList[i] = prov.PROV["location"]
        elif element == "crypto:sha512":
            termList[i] = CRYPTO["sha512"]
        elif element == "prov:value":
            termList[i] = prov.PROV["value"]
        elif element == "prov:SoftwareAgent":
            termList[i] = prov.PROV["SoftwareAgent"]            
        else:
            if element:
                found_attr = df[' Term name'] == element.split(":")[-1]

                if any(found_attr):
                    nmsp = df[found_attr].loc[:,'Namespace']
                    if len(nmsp) is not 1:
                        print "error: "+len(nmsp)
                    else:
                        namespaceName = str(nmsp.iloc[0])
                        if namespaceName == "nidm":
                            termList[i] = NIDM[element.split(":")[-1]]
                        if namespaceName == "spm":
                            termList[i] = SPM[element.split(":")[-1]]
                        if namespaceName == "fsl":
                            termList[i] = FSL[element.split(":")[-1]]                      
                else:
                    termList[i] = ""
                    comments += element+" not found"
                
    # If started with a single element return a single element     
    if not was_list:
        if termList:
            termList = termList[0]
                
    return termList, comments
```

```python
doc = prov.ProvBundle()

# Explicitely add namespaces (not included otherwise)
doc.add_namespace(NIDM)
doc.add_namespace(FSL)
doc.add_namespace(SPM)
doc.add_namespace(prov.PROV)
doc.add_namespace(ProvXSD)

for idx, series in df.iterrows():        
    
    parentFields, parentComments = updated_namespace(df, series['Parent term'])
    rangeFields, rangeComments = updated_namespace(df, series['Range'].replace(" ", "").split(","))         
    # Parse domainOrAttributes column into a list    
    domainOrAttributes, domainOrAttributesComments = updated_namespace(df, series['Domain or Attributes'].replace(" ", "").split(","))    
    
    commentTuple = zip(["Model", "Context", "Comments", "Parent", "Range", "Domain or Attributes", "BIRNLex or \nNIDM Concept ID",
                        'path', 'Database', 'Term table', 'Term column', 'Parent', 'Synonym', 'General Note', 'ExternalID', 'new parent', 
                        '"type"', 'Like', 'Comments'],
                       [series.model, series["Context"], series["Comments"], parentComments, rangeComments, 
                        domainOrAttributesComments, series["BIRNLex or \nNIDM Concept ID"], series["path"], 
                        series["Database"], series['Term table'], series['Term column'], series['Parent'], series['Synonym'], 
                        series['General Note'], series['ExternalID'], series['new parent'], series['"type"'], series['Like']])
    
    commonAttributes = [(prov.PROV['definition'], series.Definition),
                        (IAO['IAO_0000112'], series.Example),     # Example goes in "example of usage" 
                        (OWL['sameAs'], series["Synonym note"]),  # Synonym note goes in owl:sameAs
                        (RDFS['seeAlso'], series["synonyms and related urls"]),  # synonyms and related urls note goes in rdfs:seeAlso                        
                        # model and context go is "editor note" => to be processed later on                         
                        (IAO['IAO_0000116'], print_comment_tuple(commentTuple))
                        ] 
    
    commonAttributesForClasses = [(RDFS['subClassOf'], prov.PROV[series['Prov Type']]), # Parent PROV class
                                  (RDFS['subClassOf'], parentFields), # Other parent PROV class 
                                  (RDF['type'], OWL['Class'])]
    
    # Ignore empty attributes
    commonAttributes = [(a, b) for (a, b) in commonAttributes if not (not b)]  
    commonAttributesForClasses = [(a, b) for (a, b) in commonAttributesForClasses if not (not b)]      
    
    if series['Prov Type'] == 'Entity':
        doc.entity(series['Namespace']+":"+series[" Term name"], other_attributes=commonAttributes
                                                                +commonAttributesForClasses
                                                                )
    elif series['Prov Type'] == 'Activity':
        doc.activity(series['Namespace']+":"+series[" Term name"], other_attributes=commonAttributes
                                                                +commonAttributesForClasses
                                                                )
    elif series['Prov Type'] == 'Agent':
        doc.agent(series['Namespace']+":"+series[" Term name"], other_attributes=commonAttributes
                                                                +commonAttributesForClasses
                                                                )
    elif series['Prov Type'] == 'Attribute':
        doc.agent(series['Namespace']+":"+series[" Term name"], other_attributes=commonAttributes
                                    +[(RDF['type'], OWL['DatatypeProperty'])]
                                    +[(a, b) for (a, b) in zip([RDFS['domain']]* len(domainOrAttributes), domainOrAttributes) if not (not b)]
                                    +[(a, b) for (a, b) in zip([RDFS['range']]* len(rangeFields), rangeFields) if not (not b)]
                                    +[(a, b) for (a, b) in [(RDFS['subPropertyOf'], parentFields)] if not (not b)] # Other parent PROV class  
                                    )
```

```python
rdf_serialisation = doc.rdf().serialize(format='turtle')
```

```python
# Remove mention of "prov:agent" in all "DatatypeProperty" (used as convenience to write attributes using the prov toolbox)
rdf_serialisation = rdf_serialisation.replace("""a owl:DatatypeProperty,
        prov:Agent""", """a owl:DatatypeProperty""").replace("""a owl:Class,
        prov:Entity ;""", """a owl:Class ;""").replace("""a owl:Class,
        prov:Activity ;""", """a owl:Class ;""").replace("""a owl:Class,
        prov:Agent ;""", """a owl:Class ;""").replace("nidm:atCoordinateSpace a owl:DatatypeProperty", 
        "nidm:atCoordinateSpace a owl:ObjectProperty").replace("nidm:randomFieldStationarity a owl:DatatypeProperty",
        "nidm:randomFieldStationarity a owl:ObjectProperty")
rdf_serialisation += """
rdfs:subPropertyOf rdf:type rdf:Property .
rdfs:range rdf:type rdf:Property .
rdfs:domain rdf:type rdf:Property. 
prov:SoftwareAgent rdfs:subClassOf prov:Agent .
"""
print rdf_serialisation
```

```python
file = open("./data/nidm-results.ttl", "w")
file.write(rdf_serialisation)
file.close()
```
